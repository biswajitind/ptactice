# **Python Control Flow and Loops**
In this notebook, we will cover control flow statements and loops in Python, including if, if-else, if-elif, if-elif-else, loops, nested loops, break and continue statements, and functions.

## If Statements
The `if` statement is used to test a condition and execute a block of code if the condition is true.

```python
# If statement example
x = 10
if x > 5:
    print(f'{x} is greater than 5')
    print('Hello')
    print("This is being executed")

# if x == 'anshaj':
#   print("The value of x is anshaj")

# # if x == 10:
# #   print("X is 10")
```

```python
month = 'May'

if month == 'December':
  print("It's christmas time")
elif month == 'January':
  print("Its so cold")
elif month == 'February':
  print("Its still super cold")
elif month == 'March':
  print("Maybe not so cold")
else:
  print("I don't know what month it is")
```

```python
name = 'Sam'

if name == 'Anshaj':
  print("Anshaj is teaching this class")
else:
  print("You're a learner in today's session")
```

## If-Else Statements
The `if-else` statement is used to test a condition and execute one block of code if the condition is true, and another block of code if the condition is false.

```python
# If-else statement example
x = 10
if x > 15:
    print(f'{x} is greater than 15')
else:
    print(f'{x} is not greater than 15')
```

## If-Elif Statements
The `if-elif` statement is used to test multiple conditions. If the first condition is false, it checks the next condition, and so on.

```python
# If-elif statement example
x = -10
if x > 15:
    print(f'{x} is greater than 15')
elif x > 12:
  print(f'{x} is greater than 12')
elif x == 10:
  print("You've hit the jackpot")
elif x > 5:
    print(f'{x} is greater than 5 but not greater than 15')
else:
    print('x is probably less than 5')
```

## If-Elif-Else Statements
The `if-elif-else` statement is used to test multiple conditions and execute different blocks of code based on which condition is true.

```python
# If-elif-else statement example
x = 10
if x > 15:
    print(f'{x} is greater than 15')
elif x > 10:
    print(f'{x} is greater than 10 but not greater than 15')
elif x > 5:
    print(f'{x} is greater than 5 but not greater than 10')
else:
    print(f'{x} is not greater than 5')
```

## Loops
Loops are used to execute a block of code multiple times. Python supports `for` and `while` loops.

### For Loops
The `for` loop is used to iterate over a sequence (such as a list, tuple, or string) and execute a block of code for each item in the sequence.

```python
# For loop example
fruits = ['apple', 'banana', 'cherry', 'oranges', 'watermelon','grapes','pear']

for fruit in fruits:
    if fruit == 'apple':
      print("I love eating apple")
    elif fruit == 'oranges':
      print("I like drinking orange juice instead")
    elif fruit == 'banana':
      print("Yay! Banana shake")
    else:
      print("Um, I would not like to eat this fruit")

```

### While Loops
The `while` loop is used to execute a block of code as long as a condition is true.

```python
# While loop example
current_students = 0

while current_students <= 30:
    print(f'There are {current_students} students in the class')
    current_students += 1
    print("New Student Admitted")
```

```python
def smart_employee_details(**kwargs):
  mock_data = {
      1 : {"first_name": "Alex", 'occupation': 'Developer'},
      2 : {"first_name": "John", 'occupation': 'Manager'},
      3 : {"first_name": "Smith", 'occupation': 'HR'}
  }

  if kwargs.get('employee_id'):
    return mock_data.get(kwargs.get('employee_id'), "Employee not found")
  elif kwargs.get('first_name'):
    for key, value in mock_data.items():
      if value.get('first_name') == kwargs.get('first_name'):
        return value
  else:
    return "Employee not found using your parameter"
```

## Sorting in a data frame

```python
data = {
    'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04'],
    'Product': ['Product A', 'Product B', 'Product A', 'Product C'],
    'Quantity Sold': [100, 150, 120, 80],
    'Revenue': [5000, 7500, 6000, 4000]
}

# Create a DataFrame
df = pd.DataFrame(data)

df
```

```python
# See product from highest revenue to the lowest
df.sort_values(by = 'Revenue', ascending = False)
```

```python
# Sample DataFrame
data = {
    'Student_Name': ['Alice', 'Bob', 'Eve', 'David', 'Carol'],
    'Age': [25, 30, 22, 24, 28],
    'Score': [95, 93, 95, 88, 88]
}

df = pd.DataFrame(data)

df
```

```python
df.sort_values(by = 'Score', ascending = False)
```

```python
# Sort score from highest to the lowest, sort the student age from smallest to the largest
df.sort_values(by = ['Score', 'Age'] , ascending = [False, True])
```
